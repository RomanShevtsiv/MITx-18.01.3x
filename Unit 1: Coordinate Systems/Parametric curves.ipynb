# 3. Preparation

```python
# The circle
# You are familiar with the circle of radius 1, and the implicit function that defines it x^2 + y^2 = 1.

# If you are on the top half of the circle, what is the expression for y in terms of x?
x = np.linspace(-1, 1, 100)
y = np.sqrt(1 - x**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axhline(0, c='k', lw=1)
ax.axvline(0, c='k', lw=1)
ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.grid(ls='--')
plt.box(False)
plt.show()
```

```python
# If you are on the left half of the circle, what is the expression for x in terms of y?
y = np.linspace(-1, 1, 100)
x = -np.sqrt(1 - y**2)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axhline(0, c='k', lw=1)
ax.axvline(0, c='k', lw=1)
ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.grid(ls='--')
plt.box(False)
plt.show()
```

```python
# Based on the example in the video, what is a parameterization for the unit circle centered at the origin
# that starts at the point (1,0) when t = 0  and moves clockwise around the circle, returning to (1, 0)
# when t = 2*Pi?
```

```python
fig, ax = plt.subplots()

t = np.arange(0, 2*np.pi+0.1, 0.1)
line, = ax.plot(np.cos(t), -np.sin(t))
ax.axhline(0, c='k', lw=1)
ax.axvline(0, c='k', lw=1)
ax.set_aspect('equal')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
ax.grid(ls='--')
plt.box(False)

def animate(i):
    line.set_xdata(np.cos(t[:i]))  # update the data.
    line.set_ydata(-np.sin(t[:i]))  # update the data.
    return line,


ani = animation.FuncAnimation(
    fig, animate, interval=100, blit=True, save_count=100)

plt.show()
```

# 4. Eliminating parameters

```python
# Find the implicit equation for the curve given by the parametric equations x(t) = t + t^2
# and y(t) = t + 2*t^2, for t any real number.
t = np.linspace(-3, 3, 100)
x = t + t**2
y = t + 2*t**2

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axhline(0, c='k', lw=1)
ax.axvline(0, c='k', lw=1)
ax.set_aspect('equal')
ax.set_xlim(-1, 20)
ax.set_ylim(-1, 30)
ax.grid(ls='--')
plt.box(False)
plt.show()
```

```python
# Find the implicit equation for the curve given by the parametric equations
# x(t) = tan(t)
# y(t) = sec(t)
# for t in the intervals (-Pi/2,Pi/2) and (Pi/2,3*Pi/2)?

# Find the implicit equation for the curve given by the parametric equations x(t) = t + t^2
# and y(t) = t + 2*t^2, for t any real number.
t = np.linspace(-np.pi/2, 3*np.pi/2, 100)
x = np.tan(t)
y = 1/np.cos(t)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.axhline(0, c='k', lw=1)
ax.axvline(0, c='k', lw=1)
ax.set_aspect('equal')
ax.set_xlim(-1, 40)
ax.set_ylim(-1, 40)
ax.grid(ls='--')
plt.box(False)
plt.show()
```

# 5. Sketching parametric curves

```python
def calculate_xy(t: float) -> Tuple[float, float]:
    return (2*np.cos(t)-np.cos(2*t), 2*np.sin(t)-np.sin(2*t))
```

```python
for t in [0, np.pi/4, np.pi/2, np.pi]:
    print(f"t={t}: {calculate_xy(t)}")
```
